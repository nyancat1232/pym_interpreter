```python
import torch
from simpletorch.simple import TorchPlus
```

```python
class MyModel(TorchPlus):
    def process(self):
        result = self.input('input',torch.FloatTensor([1,2,3]))*self.parameter('param',torch.FloatTensor(1))
        self.label(torch.FloatTensor([3,6,9]))
        return result
#1=>3 2=>6 ... #3
```

```python
model=MyModel()
```

```python
mult_model=model.train(show_every_iteration=True)
```

```python
mult_model(input=torch.FloatTensor([5,3,2,1]))
```
