```python
import torch
from pyplus.pytorch.simple import TorchPlus
from torchvision.datasets.mnist import MNIST
```

```python
mnist=MNIST(root='./data/',download=True,)
```

```python
train=mnist.train_data.reshape(-1,28*28)
label=mnist.train_labels
```

```python
label.shape
```

```python
tp = TorchPlus()
tp.meta_data_per_iteration=100
tp.meta_epoch=10
tp.meta_optimizer=torch.optim.Adam
tp.meta_activator=torch.relu
def process(activator):
    proc = tp.input('x',train.float()) @ tp.parameter('param',torch.rand(28*28,512).float())
    proc = activator(proc)
    proc = proc @ tp.parameter('param2',torch.rand(512,512).float())
    proc = activator(proc)
    proc = proc @ tp.parameter('param3',torch.rand(512,1).float())
    
    tp.label('label',label.float())
    return proc
tp.process=process
```

```python
type(train)
```

```python
model=tp.train(show_sequence_process=True)
```

```python
test_data=mnist.test_data
test_label=mnist.test_labels
```

```python
tp.all_predict_tensors.get_tensor(name='x').tensor.shape
```
