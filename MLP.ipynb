```python
import torch
from simpletorch.simple import TorchPlus
```

```python
torch.__version__
```

```python
class MyModel(TorchPlus):
    def process(self):
        self.label(torch.zeros(3),axis_sequence=-1)
        
        x=self.input('input',torch.ones(4),axis_sequence=-1)
        
        layer_amount=[4,3]
        out=x
        for ind,val in enumerate(zip(layer_amount,layer_amount[1:])):
            out = out @ self.parameter(f'param{ind}',torch.rand(val[0],val[1]))
            out = torch.nn.functional.leaky_relu(out)
        out
        return out
```

```python
v = MyModel(meta_epoch=10000)
```

```python
func,params=v.train(show_every_iteration=True)
```

```python
func(input=torch.FloatTensor([[3,3,4,2]]))
```
