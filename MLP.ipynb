```python
import torch
from simpletorch.simple import TorchPlus,RecommendInit,MetaDataType,CurrentStateInformation
```

```python
torch.__version__
```

```python
class MyModel(TorchPlus):
    def process(self):
        self.label(torch.zeros(3),axis_sequence=-1,meta_data_type=MetaDataType.NUMERICAL)
        
        x=self.input(torch.ones(4),name='input',axis_sequence=-1,meta_data_type=MetaDataType.NUMERICAL)
        
        layer_amount=[4,3]
        out=x
        for ind,val in enumerate(zip(layer_amount,layer_amount[1:])):
            out = out @ self.parameter(size=[val[0],val[1]],init_func=RecommendInit.RELU,name=f'param{ind}')
            out = torch.nn.functional.leaky_relu(out)
        out
        return out
    def show_progress(self,csi:CurrentStateInformation):
        print(f'Epoch : {csi.current_epoch} \tIteration : {csi.current_iteration}/{csi.len_iteration}\tLoss : {csi.current_loss}')
```

```python
v = MyModel()
```

```python
func=v.train(epoch=10000)
```

```python
func(input=torch.FloatTensor([[3,3,4,2]]))
```
